How many multiplications and additions do you need to multiply a matrix between a (n, k) and (k, m) matrix? Explain.

Ans. Since the resulting matrix will have (n, m) configuration, thus nm elements, meaning nm computations. As for multiplications, the first and second matrices have k elements in each row and column. Therefore, nmk will be the total no. of multiplications. Similarly, after the k multiplications, we'll be adding them; therefore, for k multiplications, we'll need (k-1) additions. Therefore, there is a total of nm(k-1) no. of additions.


Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
# Using lists

def matrix_multiply_lists(matrix_a, matrix_b):
    result = [[0 for ele in range(len(matrix_b[0]))] for ele in range(len(matrix_a))]

    for i in range(len(matrix_a)):
        for j in range(len(matrix_b[0])):
            for k in range(len(matrix_b)):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]

    return result
```

```python
# Using NumPy

import numpy as np

def matrix_multiply_numpy(matrix_a, matrix_b):
    return np.dot(matrix_a, matrix_b)
```

NumPy is significantly faster compared to using the lists method. NumPy is highly optimized and uses efficient algorithms, resulting in faster execution.

What is the gradient of the following function with respect to x and y? (x^2)*y + (y^3)*sin x

Ans. With respect to x: 2xy + (y^3)* cos x
With respect to y: x^2 + 3y^2 sin x


```python
#Using JAX:

import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

def gradient(x, y):
    df_dx = 2 * x * y + y**3 * jnp.cos(x)
    df_dy = x**2 + 3 * y**2 * jnp.sin(x)
    return jnp.array([df_dx, df_dy])

grad_f = jax.grad(f)

# Random values for x and y
x_val = 3.0
y_val = 4.0

analytical_result = gradient(x_val, y_val)

numerical_result = grad_f(x_val, y_val)

print("Analytical Gradient:", analytical_result)
print("Numerical Gradient:", numerical_result)

```

```python
# Using sympy

import sympy as sp

x, y = sp.symbols('x y')


f = x**2 * y + y**3 * sp.sin(x)

gradient = [sp.diff(f, x), sp.diff(f, y)]

print("Analytical Gradient:", gradient)
```

Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np

def custom_argmin(arr):
    min_val = arr[0]
    min_index = 0

    for i in range(1, len(arr)):
        if arr[i] < min_val:
            min_val = arr[i]
            min_index = i

    return min_index

# Example
numpy_array = np.array([4, 2, 8, 10, 6])
result_custom = custom_argmin(numpy_array)
result_np = np.argmin(numpy_array)

print("Custom argmin result:", result_custom)
print("NumPy argmin result:", result_np)
```

Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.

Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

Ans. The time complexity of finding the median is O(n^2), where n is the array's length. A better way with the time complexity of O(n log n) is sorting and finding the median.

```python
import numpy as np
import timeit

def one_pass_method(arr):
    n = len(arr)
    for i in range(n // 2):
        max_val = max(arr)
        arr.remove(max_val)
    return arr[0] if n % 2 != 0 else (arr[0] + max(arr)) / 2

def median_sorting(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    return sorted_arr[n // 2] if n % 2 != 0 else (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2


arr = np.random.rand(1000)

one_pass_time = timeit.timeit(lambda: one_pass_method(arr.copy().tolist()), number=10)
sorting_time = timeit.timeit(lambda: median_sorting(arr.copy()), number=10)
numpy_time = timeit.timeit(lambda: np.median(arr), number=10)

# Print the results
print("One-pass method time:", one_pass_time)
print("Sorting method time:", sorting_time)
print("Numpy median time:", numpy_time)
```

Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

np.random.seed(42)
matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

max_std_column = df.std().idxmax()

min_mean_row = df.mean(axis=1).idxmin()

# Print the DataFrame and the results
print("Generated DataFrame:")
print(df)

print("\nColumn with the highest standard deviation:", max_std_column)
print("Row with the lowest mean:", min_mean_row)
```

Add a new column to the dataframe called “f” which is the sum of the columns “a”, “b”, “c”, “d”, “e”. Create another column called “g”. The value in the column “g” should be “LT8” if the value in the column “f” is less than 8 and “GT8” otherwise. Find the number of rows in the dataframe where the value in the column “g” is “LT8”. Find the standard deviation of the column “f” for the rows where the value in the column “g” is “LT8” and “GT8” respectively.

```python
import numpy as np
import pandas as pd

np.random.seed(42)
matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

max_std_column = df.std().idxmax()

min_mean_row = df.mean(axis=1).idxmin()

df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

num_lt8_rows = (df['g'] == 'LT8').sum()
std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

# Updated DataFrame
print("Updated DataFrame:")
print(df)

print("\nNumber of rows where the value in column 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of column 'f' for 'LT8' rows:", std_lt8)
print("Standard deviation of column 'f' for 'GT8' rows:", std_gt8)
```
